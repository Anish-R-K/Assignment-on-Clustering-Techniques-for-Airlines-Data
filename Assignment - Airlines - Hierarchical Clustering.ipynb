```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
```

```python
xls=pd.ExcelFile('EastWestAirlines.xlsx')
df1=pd.read_excel(xls,'data')
df1
```

```python
df1.isnull().sum()
```

```python
scaler=MinMaxScaler()
x=pd.DataFrame(scaler.fit_transform(df1.iloc[:,1:]))
x
```

```python
# Dendrogram
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch
z=linkage(x,method='complete',metric='euclidean')
plt.figure(figsize=(15,5))
sch.dendrogram(z)
plt.xlabel('Index')
plt.ylabel('Distance')
plt.title('Dendrogram')
plt.show()
```

### Dendrograms using several methods

```python
# Dendrogram
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch
z=linkage(x,method='single',metric='euclidean')
plt.figure(figsize=(15,5))
sch.dendrogram(z)
plt.xlabel('Index')
plt.ylabel('Distance')
plt.title('Dendrogram')
plt.show()
```

```python
x1=linkage(x,method='centroid',metric='euclidean')
plt.figure(figsize=(15,5))
sch.dendrogram(x1)
plt.show()
```

```python
y=linkage(x,method='average',metric='euclidean')
plt.figure(figsize=(15,5))
sch.dendrogram(y)
plt.show()
```
